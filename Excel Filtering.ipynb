```python
import openpyxl
import pandas as pd
openpyxl. __version__
from openpyxl import Workbook, load_workbook
wb = load_workbook(r"C:\Users\Downloads\ExcelDocument.xlsx")

#Prepare the Excel file in the Download folder that we need to load
df = pd.DataFrame(pd.read_excel(r"C:\Users\Downloads\ExcelDocument.xlsx"))
#Add 4 more columns in the worksheet
wb = df.reindex(columns=df.columns.tolist() + ['column1',"column2","column3","column4"])
wb["column1"] = ''
wb["column2"] = ''
wb["column3"] = ''
wb["column4"] = ''
#Rearrange and rename the columns that you would like to keep, and name workbook as Open
Open = wb[['Mapping Case', "column1", "column2", "P-Case", "Expedited", "Delegation", "Intent", "P-Case Status", "PLM_Group",
          "Assigned To", "column3", "P-Case Aging", "column4", "P-Case Date of request", "P-Case Status Date",
          "Note","P-Case Completed Date"]]
Open.rename(
    columns={"Mapping Case": "M-Case", "column1": "Total Transaction", "column2": "Total Close", "P-Case": "P-Case Open", 
            "Intent": "Work Type", "P-Case Status": "Status", "PLM_Group": "Assigned Group", "Assigned To": "PLM Assigned to",
             "column3": "In SPO?", "P-Case Aging": "AGE", "column4": "Return From PC date",
             "P-Case Date of request": "Date Received", "P-Case Status Date": "AD", "Note": "Note"
            },
    inplace=True,
)
#Filtering the types of columns, like removing those has words like 'complete', 'close', 'cancel' 
#in the status column 
Open = Open[Open["Status"].str.contains("COMPLETE") == False]
Open = Open[Open["Status"].str.contains("CLOSE") == False]
Open = Open[Open["Status"].str.contains("CANCEL") == False]
Open = Open[Open["Assigned Group"].str.contains("Credential") == False]
Open = Open[Open["Assigned Group"].str.contains("Intake Team") == False]

#Adding borders to the cells
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side 

top = Side(border_style = 'thick', color = '000000')
left = Side(border_style = 'thick', color = '000000')
right = Side(border_style = 'thick', color = '000000')
bottom = Side(border_style = 'thick', color = '000000')

border = Border(top=top, left=left, right=right, bottom=bottom)
Open.border = border
#Exporting the new document
Open.to_excel(r'C:\Users\Desktop\SavedDocument.xlsx', index=False)
```
